```python
# Function to issue a book to a student-->
def issue_book(issuedbook):
    name = input("Enter student's name: ")
    class_name = input("Enter student's class: ")
    section = input("Enter student's section: ")
    book_name = input("Enter Book name: ")
    issuedbook.append((name, class_name, section, book_name))
    print("Book issued successfully!")

# Function to view all issued books-->
def view_books(issuedbook):
    print("List of Issued Books:")
    for idx, i in enumerate(issuedbook):
        print(f"{idx}. Name: {i[0]}, Class: {i[1]}, Section: {i[2]}, Book: {i[3]}")

# Function to search for issued books by a  person-->
def search_books(issuedbook):
    search_name = input("Enter student's name to search: ")
    found = False
    print(f"Issued Books by {search_name}:")
    for index, i in enumerate(issuedbook):
        if i[0] == search_name:
            print(f"{index}. Book: {i[3]}, Class: {i[1]}, Section: {i[2]}")
            found = True
    if not found:
        print("No issued books found for this person.")
# Function to delete an Issued Book-->
def delete_book(issuedbook):
    view_books(issuedbook)
    if issuedbook:
            index = int(input("Enter the index of the book you want to delete: "))
            if 1 <= index <= len(issuedbook):
                del issuedbook[index]
                print("Book deleted successfully!")
            else:
                print("Invalid index!")
    else:
        print("No books to delete.")

# Function to save issued books data to a file using file handling-->
def save_books(issuedbook):
    with open("issuedbook.txt", "w") as file:
        for i in issuedbook:
            file.write(f"{i[0]},{i[1]},{i[2]},{i[3]}\n")
    print("Issued books data saved successfully!")

# Function to load issued books data from a file-->
def load_books():
    issuedbook = []
    try:
        with open("issuedbook.txt", "r") as file:
            for line in file:
                issuedbook.append((name, class_name, section, book_name))
    except:
        print("No existing data found.")
    return issuedbook

# Main function to run the program-->
def main():
    issuedbook = load_books()
    while True:
        print("\nIssued Books Management")
        print("1. Issue Book")
        print("2. View Issued Books")
        print("3. Search Issued Books by Person")
        print("4. Delete")
        print("5. Exit")
        choice = input("Enter your choice: ")
        if choice == '1':
            issue_book(issuedbook)
        elif choice == '2':
            view_books(issuedbook)
        elif choice == '3':
            search_books(issuedbook)
        elif choice == '4':
            delete_book(issuedbook)
        elif choice == '5':
            save_books(issuedbook)
            print("Exiting...")
            break
        else:
            print("Invalid choice!")

if __name__ == "__main__":
    main()
```
