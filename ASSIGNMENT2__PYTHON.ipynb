```python
import pandas as pd

class Weather:
    def __init__(self, city, temperature, humidity, wind_speed):
        self.city = city
        self.temperature = temperature
        self.humidity = humidity
        self.wind_speed = wind_speed

    def display(self):
        print(f"Weather in {self.city}:")
        print(f"Temperature: {self.temperature}°C")
        print(f"Humidity: {self.humidity}%")
        print(f"Wind Speed: {self.wind_speed} km/h")

class WeatherManage:
    def __init__(self, filename='weather_data.txt'):
        self.filename = filename

    def add_weather(self, weather):
        with open(self.filename, 'a') as file:
            file.write(f"{weather.city},{weather.temperature},{weather.humidity},{weather.wind_speed}\n")

    def view_weather(self):
        try:
            weather_data = pd.read_csv(self.filename, header=None, names=['City', 'Temperature (°C)', 'Humidity (%)', 'Wind Speed (km/h)'])
            print(weather_data)
        except FileNotFoundError:
            print("No weather data available.")

    def delete_weather(self, city):
        try:
            with open(self.filename, 'r') as file:
                lines = file.readlines()
            with open(self.filename, 'w') as file:
                for line in lines:
                    if city not in line:
                        file.write(line)
            print(f"Weather data for {city} deleted successfully!")
        except FileNotFoundError:
            print(f"Weather data file not found.")

def add_weather():
    city = input("Enter city name: ")
    temperature = float(input("Enter temperature in Celsius: "))
    humidity = float(input("Enter humidity percentage: "))
    wind_speed = float(input("Enter wind speed in km/h: "))

    return Weather(city, temperature, humidity, wind_speed)

def main():
    manage = WeatherManage()

    while True:
        print("\nMenu:")
        print("1. Add Weather Data")
        print("2. View Weather Data")
        print("3. Delete Weather Data")
        print("4. Exit")
        choice = input("Enter your choice: ")

        if choice == '1':
            try:
                weather = add_weather()
                manage.add_weather(weather)
                print("Weather data added successfully!")
            except ValueError:
                print("Invalid input! Please enter numeric values for temperature, humidity, and wind speed.")

        elif choice == '2':
            manage.view_weather()

        elif choice == '3':
            city = input("Enter the city name to delete: ")
            manage.delete_weather(city)

        elif choice == '4':
            print("Exiting the program...")
            break

        else:
            print("Invalid choice! Please enter a valid option.")

if __name__ == "__main__":
    main()
```
